```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/titanic_train.csv')
tit = df.copy()
```

```python
df.head()
```

```python
df.drop('PassengerId',axis=1,inplace=True)
df.head()
```

```python
del df['Name']
df.head()
```

```python
df.drop(['Ticket','Parch'],axis=1,inplace=True)
df.head()
```

```python
df.drop(['SibSp','Fare'],axis=1,inplace=True)
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
def sex(s):
  if s=='male':
    return 0
  elif s=='female':
    return 1


df['Gender'] = df.Sex.apply(sex)

df.head()
```

```python
del df['Sex']
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.Embarked.fillna(df.Embarked.mode()[0], inplace=True)
```

```python
del df['Cabin']
```

```python
df.describe()
```

```python
def embarked(s):
  if s=='S':
    return 0
  elif s=='C':
    return 1
  elif s=='Q':
    return 2


df['Embarked']=df.Embarked.apply(embarked)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
mean_age_survived = df[df['Survived'] == 1]['Age'].mean()
mean_age_deceased = df[df['Survived'] == 0]['Age'].mean()


def fill_age(row):
  if pd.isnull(row['Age']):
    if row['Survived']==1:
      return mean_age_survived
    else:
      return mean_age_deceased
  else:
    return row['Age']

df['Age'] = df.apply(fill_age, axis=1)
```
